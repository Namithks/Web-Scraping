```python
# Step 1: Import necessary libraries
import matplotlib.pyplot as plt

# Total 18 products
categories = [
    "Semiconductors", "Embedded Systems", "Displays", "Micro- and Nanosystems",
    "Sensor Technology", "Test and Measurement", "Electronic Design (ED/EDA)",
    "Passive Components", "Electromechanics", "Power Supplies",
    "PCBs and Other Circuit Carriers", "EMS Electronic Manufacturing Services",
    "Assemblies and Subsystems", "Automotive", "Wireless", "Services",
    "Printed Flexible Electronics", "SEMICON Europa"
]

# Corresponding product counts
product_counts = [
    1623, 553, 415, 207,
    562, 934, 157,
    1742, 3843, 1136,
    1061, 1024, 165,
    545, 390, 484,
    174, 492
]


plt.figure(figsize=(12, 6))
plt.barh(categories, product_counts, color='skyblue')
plt.xlabel('Number of Products')
plt.title('Number of Products by Category')
plt.gca().invert_yaxis()  # Invert y-axis to have the first category on top
plt.grid(axis='x')

# For Display the graph
plt.show()
```
