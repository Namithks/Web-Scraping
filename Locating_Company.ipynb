```python
import folium

# Initialize map
m = folium.Map(location=[48.13635160106622, 11.6929703488861], zoom_start=13)

# Add a marker with a building icon and custom color
folium.Marker(
    location=[48.13635160106622, 11.6929703488861],
    popup="Electronica Center Munich",
    icon=folium.Icon(color='green', icon="fa-building", prefix="fa")
).add_to(m)

m
```
